```python
#handling missing values
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='median')
df1[['Age']] = imp.fit_transform(df1[['Age']])
df1
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='mean')
df1[['Salary']] = imp.fit_transform(df1[['Salary']])
df1
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='most_frequent')
df1[['Gender']] = imp.fit_transform(df1[['Gender']])
df1
```

```python
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
iqr = q3-q1
lower_bound = q1-1.5*iqr
upper_bound = q3+1.5*iqr

outlier= df[(df['Age']<lower_bound) | (df['Age']>upper_bound )]
outlier
```

```python
from scipy.stats import zscore
z_scores = zscore(df['Age'])
df[np.abs(z_scores)>2.5] #common rule : z>3 means outlier
```

```python
from scipy.stats import zscore
z_scores = zscore(df['Salary'])
df[np.abs(z_scores)>2.5]
# gender is categorical so need to encode it
```

```python
from sklearn.ensemble import IsolationForest
iso = IsolationForest(contamination=0.1)
df['outlier'] = iso.fit_predict(df[['Age', 'Salary']])
df[df['outlier'] == -1]  # -1 means outlier
#if you give contamination value more it give more value like 0.2,..
```

```python
from sklearn.ensemble import IsolationForest
iso = IsolationForest(contamination=0.3)
df['outlier'] = iso.fit_predict(df[['Age', 'Salary']])
df[df['outlier'] == -1]
```

```python
#winsorization fv
from scipy.stats.mstats import winsorize

df1['Age_wins'] = winsorize(df1['Age'], limits=[0, 0.1])# here the limit imdicates 0 to 10% any value can be given

df1['Salary_wins'] = winsorize(df1['Salary'], limits=[0, 0.1])

#df1['Gender'] = winsorize(df1['Gender'], limits=[0, 0.1])
print(df1)
```

```python
import numpy as np
df2['Age'].fillna(df2['Age'].mean(), inplace=True)
df2['Salary'].fillna(df2['Salary'].mean(), inplace=True)
df2['Age_log'] = np.log1p(df2['Age'])
df2['Salary_log'] = np.log1p(df2['Salary'])
print(df2)
```

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='yeo-johnson')#this method handles both +,_and 0 but Box-Cox method  handles only +

df2[['Age', 'Salary']] = pt.fit_transform(df2[['Age', 'Salary']])

print(df2)
```

```python
from sklearn.preprocessing import StandardScaler
df3['Age'].fillna(df3['Age'].mean(), inplace=True)
df3['Salary'].fillna(df3['Salary'].mean(), inplace=True)
scaler = StandardScaler()

df3[['Age', 'Salary']] = scaler.fit_transform(df3[['Age', 'Salary']])

print(df3)
```

```python
from sklearn.preprocessing import MinMaxScaler

df3['Age'] = df3['Age'].fillna(df3['Age'].mean())
df3['Salary'] = df3['Salary'].fillna(df3['Salary'].mean())

scaler = MinMaxScaler()

df3[['Age', 'Salary']] = scaler.fit_transform(df3[['Age', 'Salary']])
print(df3)
```

```python
from sklearn.preprocessing import PowerTransformer

df3['Salary'] = df3['Salary'].fillna(df3['Salary'].mean())

pt = PowerTransformer(method='box-cox')

df3[['Salary_BoxCox']] = pt.fit_transform(df3[['Salary']] + 1)

print(df3)
```
